```python
import pandas as pd
import numpy as np

import keras
from keras.models import Sequential
from keras.layers import Dense
```

```python
from keras.optimizers import Adam
from sklearn.model_selection import train_test_split
```

```python
concrete_data = pd.read_csv('https://s3-api.us-geo.objectstorage.softlayer.net/cf-courses-data/CognitiveClass/DL0101EN/labs/data/concrete_data.csv')
concrete_data.head()
```

```python
concrete_data.describe()
```

```python
concrete_data.isnull().sum()
```

```python
concrete_data_columns = concrete_data.columns

x = concrete_data[concrete_data_columns[concrete_data_columns != 'Strength']] # all columns except Strength
y = concrete_data['Strength'] # Strength column
```

```python
x.head(2)
```

```python
y.head(2)
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3)

print('x_train:', x_train.shape, 'y_train:', y_train.shape)
print('x_test:', x_test.shape, 'y_test:', y_test.shape)
```

```python
n_cols = x_test.shape[1]
```

# Baseline Model

```python
def regression_model():
    # create model
    model = Sequential()
    model.add(Dense(10, activation='relu', input_shape=(n_cols,)))
    model.add(Dense(1))
    
    # compile model
    model.compile(optimizer='adam', loss='mean_squared_error', )
    return model
```

```python
model = regression_model()

model.fit(x_train, y_train, validation_data=(x_test, y_test), epochs=50, verbose=2)

scores = model.evaluate(x_test, y_test, verbose=2)

scores
```

The mean square error is 716.1646888510695

# Model Using Normalized Data

```python
# Normalize data

x_train_norm = (x_train - x_train.mean()) / x_train.std()
x_test_norm = (x_test - x_test.mean()) / x_test.std()

x_train_norm.head(2)
```

```python
x_test_norm.head(2)
```

```python
n_cols = x_test_norm.shape[1]
n_cols
```

```python

def regression_model():
    # create model
    model = Sequential()
    model.add(Dense(10, activation='relu', input_shape=(n_cols,)))
    model.add(Dense(1))
    
    # compile model
    model.compile(optimizer='adam', loss='mean_squared_error', )
    return model
```

```python

model = regression_model()


model.fit(x_train_norm, y_train, validation_data=(x_test_norm, y_test), epochs=50, verbose=2)

scores = model.evaluate(x_test_norm, y_test, verbose=2)

scores
```

The error using the nomalized data is 506.5680151880752 and it is lower when compared to the error obatained when is data is not nomalized (716.1646888510695)

# Increasing the Number of Epoch

```python
def regression_model():
    # create model
    model = Sequential()
    model.add(Dense(10, activation='relu', input_shape=(n_cols,)))
    model.add(Dense(1))
    
    # compile model
    model.compile(optimizer='adam', loss='mean_squared_error', )
    return model
```

```python
model = regression_model()


model.fit(x_train_norm, y_train, validation_data=(x_test_norm, y_test), epochs=100, verbose=2)

scores = model.evaluate(x_test_norm, y_test, verbose=2)

scores
```

The error obtained when the numbr of epochs is increased to 100 is 155.73840183887665 and it is lower than that obtained when 50 epoch was used (506.5680151880752)

# Increased the Number of Hidden Layers

```python
def regression_model():
    # create model
    model = Sequential()
    model.add(Dense(10, activation='relu', input_shape=(n_cols,)))
    model.add(Dense(10, activation='relu'))
    model.add(Dense(10, activation='relu'))
    model.add(Dense(1))
    
    # compile model
    model.compile(optimizer='adam', loss='mean_squared_error', )
    return model
```

```python
model = regression_model()


model.fit(x_train_norm, y_train, validation_data=(x_test_norm, y_test), epochs=50, verbose=2)

scores = model.evaluate(x_test_norm, y_test, verbose=2)

scores
```

Compared to the error obtained when 50 epochs was used (506.5680151880752); when the number of hidden layers is increased to 3, the error obtaied is lesser with a value of 125.23856198209003
